```python
from IPython.display import Audio
import scipy.io.wavfile
Fs, y = scipy.io.wavfile.read('train_bird.wav')
Audio(y, rate=Fs)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y)
plt.title("Orignial sound: The combined time-domain signal of the train and bird sounds")
```

```python
import numpy as np
from numpy.fft import *
DFT_Y = fft(y)
DFT_bird = fft(y)
DFT_train = fft(y)

plt.plot(np.abs(DFT_Y))
plt.title("Frequency Domain: DFT of Orginal Sound")
```

```python

#filter bird and train freq
DFT_bird = []
DFT_train = []

i =0;
total_len = len(DFT_Y)
for Y in DFT_Y:
    if( (i < 3500) or ((total_len - 3500) < i)):  #train signal
        DFT_bird.append(0)
        DFT_train.append(Y)
    else:
        DFT_bird.append(Y)
        DFT_train.append(0)
    i = i+1
    
    
IDFT_bird = ifft(DFT_bird).real
IDFT_train = ifft(DFT_train).real
```

```python
Audio(IDFT_bird,rate=Fs)
```

```python
Audio(IDFT_train,rate=Fs)
```

```python
#plot all graphs 
plt.plot(np.abs(DFT_bird))
plt.title('Bird DFT: Filtered Frequency Domain ')
```

```python
plt.plot(np.abs(DFT_train))
plt.title('Train DFT: Filtered Frequency Domain')
```

```python
plt.plot(IDFT_bird)
plt.title('Bird IDFT: filtered time-domain signal')
```

```python
plt.plot(IDFT_train)
plt.title('Train IDFT: filtered time-domain signal')
```
